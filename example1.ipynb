```python
from tools import *
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def cv2_imshow(img):
    plt.figure(figsize=(8,5))
    plt.imshow(img[:,:,::-1])
```

```python
img = cv2.imread("images/lay.jpg")
cv2_imshow(img)
```

```python
faceDet = face_detector()
faceFeature = face_features()
```

```python
faceDet(img)
```

```python
img_2 = img.copy()
for bbox in faceDet(img,ret="bb"):
    cv2.rectangle(img_2, *bbox, (0, 0, 255), 2)
cv2_imshow(img_2)
```
